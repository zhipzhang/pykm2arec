```python
from pykm2arec import KM2AReconstructor
reconstructor = KM2AReconstructor(array_flag=7)
```

```python
source = KM2AEventSource("KM2AMCTest.root")

with KM2ARecWriter(path="rec.root") as writer:
    for event in source:
        rec_event = reconstructor(event)
        writer(rec_event)
        
```
